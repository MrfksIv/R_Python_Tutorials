### Accessing a dataframe ###
Several functions exist that allow us to access specific parts of a dataframe: __.loc[]__, __.iloc[]__ or simply __[]__. The simple square bracket notation expects ['column']['row']

```python
am_201705['barcode'][600011]
#
```

```python
# Double square brackets can be used to select multiple columns and ensure that the result is a pandas.dataframe:
am_201705[['barcode','salesdate']].head()
```

```python
# The easiest way to subselect columns is using the .loc & .iloc accessors: Their syntax is the same, but iloc expects
# index numbers while .loc expects column names:

am_201705.loc[[600011]] # This returns a dataframe
am_201705.loc[600011] # This returns a series

# A Series is 1-dimensional array with a labelled index like a hubrid between a np.array and a dictionary.
```

```python
am_201705.loc[600011, 'salesdate']
```

```python
# Note that .loc can also use colon slicing. UNLIKE 'NORMAL' colon slicing, the last value is included!
am_201705.loc[600011:600012, 'barcode':'loyaltyno']
```

```python
# .iloc allows us to to use array-slicing notation:
am_201705.iloc[0:3, :] # Selects first three rows, and all of columns
```

```python
# both .loc and .iloc can take lists instead of colon slicing:
am_201705.loc[600011:600012, ['barcode', 'loyaltyno', 'salesdate', 'salestime']]
```

### Filtering Dataframes ###
We can use multiple filtering using logical operations in the following way:

```python
am_201705[(am_201705['salesvalue'] > 4) & (am_201705['quantity'] > 2)].head()
```

```python
am_201705[(am_201705['salesvalue'] > 20) | (am_201705['salesvalue'] < 0.05)].head()
```

```python
# Returns a boolean series with index of the original columnnames which is True if ALL elements are non-zero and False otherwise.
# For example, because some rows' salesvalue, price, etc. below are 0, their value is False
am_201705.all()
```

```python
# .any() returns True of there exists at least one non-zero value. It seems from the result below that the two price_columns
# are always zero.
am_201705.any()
```

```python
# Indeed, they are always 0 since no columns are returned here.
am_201705[am_201705['price']>0]
```

```python
# We can combine these functions with .loc[] to drop for example the all-zero columns:

am_201705 = am_201705.loc[:, am_201705.any()]
```

```python
am_201705.columns
# Sometimes, instead of 0s, dataframes contain NaNs. in these cases we can use the .isnull() and notnull() 
# in combination with .any() and .all()
#
# df.isnull().any()  -> returns columns that contain at least one NaN value
# df.notnull().all() -> returns columns that contain only NON-NaN values
```

```python
# We can drop NaN rows using the method 
am_201705.dropna(how='any') # <- if any one value of the row is NaN, the row is dropped
am_201705.dropna(how='all') # <- if all values of the row are NaNs, the row is dropped

am_201705.dropna(thresh=1000, axis='columns') # <- drop columns that have more than 1000 empty vals.
```

### Transforming Dataframes ###
There exist different ways of transforming dataframes such as adding columns, creating new columns from previous ones, etc.

```python
am_201705.head()
```

```python
storecode_dict = {'600011': 'Egkomi', '600012': 'Aglantzia', '600015': 'Limassol'}
am_201705['storename'] = am_201705['storecode'].map(storecode_dict)
am_201705.head()
```

```python
# We can add columns based on calculation of other columns:

am_201705['price'] = am_201705['salesvalue'] / am_201705['quantity']
am_201705.head()

```

```python
# or using the map() function. Suppose we want to apply a discount of 20%

am_201705['salesvalue_discounted'] = am_201705['salesvalue'].map(lambda x : x * .8)
am_201705.head()
```
