```python
import pandas as pd
```

```python

# 

data1 = {
    'months': ['Jan', 'Feb', 'Mar', 'Apr'],
    'values': [3, 4, 2, 12]
}

data2 = {
    'months': ['May', 'Jun', 'Jul', 'Aug'],
    'values': [11, 19, 22, 30]
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
```

```python
df1
```

```python
# We can append one dataset below the other using the append() method.
# The append() method only appends vertically. The concat() method can do it in both directions.
df = df1.append(df2)
df
```

```python
# Notice that above, the index is not changed and the two indexes are also appended.
# This has the effect of:
df.loc[3]
```

```python
# To make the index recalculate itself, we use reset_index()
df= df1.append(df2).reset_index()
df
```

```python
# Notice above that the previous index became a column:
df.columns
```

```python
# To drop it we pass the argument drop=True to the reset_index() method:
df = df1.append(df2).reset_index(drop=True)
df
```

```python
df = df1.append(df2, ignore_index=True)
df
```

```python
# Alternatively, we can use pd.concat() method:
df = pd.concat([df1, df2], ignore_index=True)
df
```
