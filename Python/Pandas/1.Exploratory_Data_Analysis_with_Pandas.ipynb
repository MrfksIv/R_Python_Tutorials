```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
col_names = ['sepal_length','sepal_width','petal_length','petal_width','species']

iris = pd.read_csv("data/iris.txt", index_col=False, names=col_names)
iris.head()
```

```python
iris.index
```

```python
iris.plot(x='sepal_length', y='sepal_width')
plt.show()
```

The above plot doesn't make any sense because by default, pyplot makes a line-chart which for unordered data doesn't make much sense.

```python
iris.plot(x='sepal_length', y='sepal_width', kind='scatter')
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')

plt.show()
```

```python
iris.plot(y='sepal_width', kind='box')
plt.ylabel('sepal width (cm)')
plt.show()
```

```python
iris.plot(y='sepal_length', kind='hist')
plt.xlabel('sepal length (cm)')
plt.show()
```

```python
# From the above we cannot really decipher the shape of the distribution is bell-shaped or none. 
# This is because the bins are too wide. Try the following:

iris.plot(y='sepal_length', kind='hist', bins=30, range=(4,8), normed=True)
plt.xlabel('sepal length (cm)')
plt.show()
```

```python
# This formats the plots such that they appear on separate rows
fig, axes = plt.subplots(nrows=2, ncols=1)

# Plot the PDF
iris.sepal_length.plot(ax=axes[0], kind='hist', bins=30, normed=True, range=(4,8))
plt.show()

# Plot the CDF
iris.sepal_length.plot(ax=axes[1], kind='hist', bins=30, normed=True, cumulative=True, range=(4,8))
plt.show()
```

```python
# We can summarize dataframes with .describe() method of the pd.dataframe:
iris.describe()
```

```python
# Summary statistic methods can be called individually:
iris.std()
```

```python
# And even on individual columns:
iris['sepal_length'].std()
```

```python
# Notice that the species is a categorical variable:
iris.species.describe()

# count  :=  number of non-null entries
# unique :=  number of distinct values
# top    :=  most frequent category
# freq   :=  occurences of most frequent category (top)
```

```python
# We can get the distinct values of a categorical columns using the .unique() method:
iris.species.unique()
```

```python
# We can filter the dataframe based on the value of a categorical column in the following way:

indeces_to_keep = iris['species'] == 'Iris-setosa' 
setosa = iris.loc[indeces_to_keep, :] # Extracts new dataframe

setosa.head()
```

```python
# These statistics can be calculated for groups of categorical variables using the groupby() method
# before calling the summary statistic method:
iris.groupby('species').describe()
```
